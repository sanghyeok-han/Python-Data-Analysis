```python
import os
from os import listdir
from os.path import isfile, join
import re
import pandas as pd
import numpy as np
```

```python
base_path = r"C:\Users\User\Desktop\주피터 노트북 작업 공간\프로젝트\정보디자인"
all_file_names = [f for f in listdir(base_path) if isfile(join(base_path, f))]
all_file_names = [file_name for file_name in all_file_names if file_name[-3:] == "csv"]
all_file_names
```

```python
cumulated_df = pd.read_csv(all_file_names[0])

for file_name in all_file_names[1:]:
    new_df = pd.read_csv(file_name)
    cumulated_df = pd.concat([cumulated_df, new_df], axis=0)
```

```python
cumulated_df.to_csv("조선왕조실록.csv", index=False, encoding="utf-8-sig")
```
